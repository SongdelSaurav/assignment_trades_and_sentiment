```python
#code is being cleaned and prepared the sentiment dataframe for merging
sentiment_df.columns = sentiment_df.columns.str.lower()
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])

sentiment_df.rename(columns={'classification': 'sentiment'}, inplace=True)
```

```python
#code is being cleaned and prepared the trades dataframe for merging
trades_df.columns = trades_df.columns.str.lower().str.replace(' ', '_')

trades_df['timestamp_ist'] = pd.to_datetime(trades_df['timestamp_ist'])
trades_df['date'] = trades_df['timestamp_ist'].dt.date

 
```

```python
#converting the 'date' columns in the same format
trades_df['date'] = pd.to_datetime(trades_df['date']).dt.date
sentiment_df['date'] = pd.to_datetime(sentiment_df['date']).dt.date
```

```python
#checking if they match or not
print("Trades date type:", type(trades_df['date'].iloc[0]))
print("Sentiment date type:", type(sentiment_df['date'].iloc[0]))
print("\nTrades date sample:", trades_df['date'].head())
print("Sentiment date sample:", sentiment_df['date'].head())
```

```python
# merging both the datasets
merged_df = trades_df.merge(
    sentiment_df[['date', 'sentiment']],
    on='date',
    how='left'
)

print("\nMerge successful! Shape:", merged_df.shape)
print("Merged columns:", merged_df.columns.tolist())
print("Null sentiments:", merged_df['sentiment'].isnull().sum())
```

```python
print(merged_df.columns.tolist())
```

```python
# Trade outcome
merged_df['is_profitable'] = merged_df['closed_pnl'] > 0

# Absolute position size
merged_df['abs_size'] = merged_df['size_tokens'].abs()
```

```python
#performance analysis by market sentiment
performance = merged_df.groupby('sentiment').agg(
    total_trades=('account', 'count'),
    win_rate=('is_profitable', 'mean'),
    avg_pnl=('closed_pnl', 'mean'),
    avg_leverage=('size_usd', 'mean'),
    avg_position_size=('abs_size', 'mean')
).reset_index()

performance['win_rate'] = performance['win_rate'] * 100
performance
```

```python
#trade volume by sentiment
plt.figure()
plt.bar(performance['sentiment'], performance['total_trades'])
plt.title("Trade Count by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")
plt.show()
```

```python
#average PnL by sentiment
plt.figure()
plt.bar(performance['sentiment'], performance['avg_pnl'])
plt.title("Average PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average PnL")
plt.show()
```

```python
#win rate comparission
plt.figure()
plt.bar(performance['sentiment'], performance['win_rate'])
plt.title("Win Rate (%) by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Win Rate (%)")
plt.show()
```

```python
plt.figure()
plt.bar(performance['sentiment'], performance['avg_leverage'])
plt.title("Average Leverage by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Leverage")
plt.show()
```
